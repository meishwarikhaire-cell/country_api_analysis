```python
#Task 1 — Fetch and Validate the Response

import requests

url = "https://restcountries.com/v3.1/all"

params = {
    "fields": "name,population,region,area"
}

try:
    response = requests.get(url, params=params)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries:", len(data))
    else:
        print("Error:", response.text)

except requests.exceptions.RequestException as e:
    print("Request Error:", e)
```

```python
#Task 2 — Extract and Display Key Fields
import pandas as pd

# Extract required fields
country_list = []

for country in data:
    country_info = {
        "common_name": country.get("name", {}).get("common"),
        "population": country.get("population"),
        "region": country.get("region"),
        "area": country.get("area")
    }
    country_list.append(country_info)

# Create DataFrame
df_countries = pd.DataFrame(country_list)

# Display first 10 rows
df_countries.head(10)
```

```python
#Task 3 — Summarize the Data

# 1. Region with highest total population
region_population = df_countries.groupby("region")["population"].sum()
highest_region = region_population.idxmax()

# 2. Country with largest area
largest_country = df_countries.loc[df_countries["area"].idxmax(), "common_name"]

# Print results
print("Region with Highest Total Population:", highest_region)
print("Country with Largest Area:", largest_country)
```
